#Two Plots In The Same Graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
dsa_marks = [90,92,96,99]
ml_marks = [91,93,95,97]

plt.plot(x,dsa_marks)
plt.plot(x,ml_marks)

plt.title("Marks corresponding to no. of hrs studied")

plt.xlabel("no. of hrs studied")
plt.ylabel("Marks")

plt.xticks([1,2,3,4])
plt.yticks([90,92,94,96,100])

plt.show()
```

# Now we have to see that which plot contains which data for this we Use "Legend

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
dsa_marks = [90,92,96,99]
ml_marks = [91,93,95,97]

plt.plot(x,dsa_marks, label = 'dsa marks')
plt.plot(x,ml_marks, label = 'ml marks')

plt.title("Marks corresponding to no. of hrs studied")

plt.xlabel("no. of hrs studied")
plt.ylabel("Marks")

plt.xticks([1,2,3,4])
plt.yticks([90,92,94,96,100])

plt.legend(loc="lower right")

plt.show()
```

#Axes, Generation of Subplots

```python
# axes - x axis, y axis, labels, tickers

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, 100)

fig, axes = plt.subplots(2, 3)

axes[0,0].plot(x, np.sin(x))
axes[0,0].set_title('Sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')

axes[0,1].plot(x, np.cos(x))
axes[0,1].set_title('Cos func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x, np.tan(x))
axes[0,2].set_title('Tan func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x, np.exp(x))
axes[1,0].set_title('Exp func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x, np.log(x))
axes[1,1].set_title('Log func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x,x)
axes[1,2].set_title('Linear Function')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')

plt.tight_layout()
```

# Cumulative Histogram

```python
#Cumulative Means It adds up the value from the previous bin

import matplotlib.pyplot as plt

ages = [34, 23, 78, 54, 67, 45, 21, 98, 54, 76]

sorted_ages = np.sort(ages)
print(sorted_ages)

plt.hist(ages, bins=5, orientation='horizontal',hatch='o', edgecolor = 'Black')
plt.title("Marks corresponding to no. of hrs studied")
plt.xlabel("no. of hrs studied")
plt.ylabel("Marks")

plt.show()
```

#3D  Bar Plots

```python
ax = plt.axes(projection="3d")

x = np.random.randint(1, 10, 5)
y = np.random.randint(1, 10, 5)
z = np.zeros(5)

dx = np.ones(5)
dy = np.ones(5)
dz = np.random.randint(1, 10, 5)

print(x, y, z, dx, dy, dz)

ax.bar3d(x, y, z, dx, dy, dz)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


plt.show()
```

# 3D Plot Surface

```python
ax = plt.axes(projection="3d")

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5,0.25)
X, Y = np.meshgrid(X,Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, cmap='plasma')

plt.show()
```
